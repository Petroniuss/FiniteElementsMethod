Finite Elements Method

```scala
import $ivy.`iet::finite-elements-method:0.1`
import agh.iet.devs.{FiniteElementsMethod, Functions}
import agh.iet.devs.Functions._
import scala.math

import $ivy.`org.plotly-scala::plotly-almond:0.7.3`

import plotly._
import plotly.element._
import plotly.layout._
import plotly.Almond._
```

```scala
val a = Functions.const(1)
val b: RealFunction = x => 0
val c: RealFunction = x => 0
val f: RealFunction = x => 20 * math.sin(10 * math.Pi * x)
val beta = 0
val gamma = 0
val uR = 2
val n = 100

val u = FiniteElementsMethod.solve(a, b, c, f) (beta, gamma, uR, n)
```

```scala
{{  
    
  var xs = Seq[Double]()
  var ys = Seq[Double]()
    
  val n = 100
    
  for (i <- 0 to n) {
      xs = xs :+ i.toDouble / n
      ys = ys :+ u(i.toDouble / n)
  }  
    
  val trace = Scatter(
    xs,
    ys,
    fill = Fill.ToNextY
  )

  val data = Seq(trace)

  plot(data)
}}
```
